```python
import pandas as pd

df = pd.read_csv('./Data/recommender_data.csv')

def recommender(selection):
#     determine if selection is a whiskey or beer
    x = df.loc[df['name'] == selection, 'id'].iloc[0]
    
#     sorting on the selected beer/whiskey
    df_sorted = df.sort_values(selection, ascending=False)
    
#     droping all opposite styles
    df_target = df_sorted[df_sorted['id'] != x]
    
#     print recommendation
    print(f'Your top 3 recommedations based on {selection} are:\n 1. {df_target["name"].unique()[0]} \n 2. {df_target["name"].unique()[1]} \n 3. {df_target["name"].unique()[2]}')
```
